```python
# import
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
#from sklearn.datasets import load_wine

# prep dataframe
data = pd.read_csv('DelinquentBusinessLicenses.csv', usecols=[1, 6, 7, 8, 9, 11, 12, 13, 14, 16, 17, 18, 19, 20, 21, 23, 24, 25])
```

```python
data.columns
```

```python
# prep dataframe

df = pd.DataFrame(
    data=data["START DATE"],
    columns=data["COMPANY TYPE"]
)
df["#EMPLOYEES"] = data["#EMPLOYEES"]

# inspect df
df.head()
```

```python
classes, counts = np.unique(
    data["PARTNER"],
    return_counts=True
)

# actual plot
for (cls, cnt) in zip(classes, counts):
    plt.bar(
        x=cls,
        height=cnt,
        label="COMPANY TYPE {cls}".format(**locals())
    )
plt.legend(loc=0)
plt.xticks([1, 2, 3])
plt.title("Count by COMPANY TYPE");
```
