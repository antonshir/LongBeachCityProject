```python
def format_query(limit, offset, hood, term):
    hood = hood + ' long beach'
    query = gql('''
    {
      search(term: "''' + str(term) + '''",
         location: "''' + str(hood) + '''",
  		 limit: ''' + str(limit) + ''',
         offset: ''' + str(offset) + ''' ) {
            total
            business {
              id
              name
              review_count
              photos
              url
              categories {
                title
                alias
              }
              coordinates {
                latitude
                longitude
              }
              location {
                address1
                city
                state
                country
                postal_code
              }
            }
        }
    }

    ''')
    return query
    
```

```python
def get_line(input):
    line_composition = []
    
    line_composition.append(input['id'])
#    print(input['id'])
    line_composition.append(input['name'])
#    print(input['name'])
    line_composition.append(format_address(input))
 #   print(format_address(input))

#    print(input['coordinates'])
#    print(input['coordinates']['longitude'])
    

          
    if ((input['coordinates']) != None) and (input['coordinates']['longitude']) != None:
        line_composition.append(str(input['coordinates']['longitude']))
#        print(input['coordinates']['longitude'])
    else:            
        line_composition.append('None')
        

    if ((input['coordinates']) != None) and (input['coordinates']['latitude']) != None:
        line_composition.append(str(input['coordinates']['latitude']))
    #        print(input['coordinates']['longitude'])
    else:
        line_composition.append('None')
            
    if (input['url']) != None:   
        line_composition.append(str(input['url']))
    else:
        line_composition.append('None')
        
    if (input['photos'][0]) != None:
        line_composition.append(str(input['photos'][0]))
    else:
        line_composition.append('None')
        
    if (format_category(input)) != None:
        line_composition.append(format_category(input))
    else:
        line_composition.append('None')
        
    if (str(input['review_count'])) != None:
        line_composition.append(str(input['review_count']))
    else:
        line_composition.append('None')
        
    if ((input['location']) != None) and (input['location']['postal_code']) != None:
        line_composition.append(str(input['location']['postal_code']))
#        print(input['coordinates']['longitude'])
    else:            
        line_composition.append('None')
        
    line_composition.append("Outdoors")
       
    return line_composition

def format_address(input):
    address = ''
    
    for line in input['location']:
        address += str(input['location'][line]) + ' '
    
    return address

def format_category(input):
    category = ''
    
    for line in input['categories']:
        category += str(line['title']) + ' '
        category += str(line['alias']) + ' '
        
    return category
```

```python
def offset_setter(offset, business_num):
    iterator = 50
    business_num = business_num
    offset = offset
    limit = 950
    
    if (offset + iterator) > limit:
        offset = limit
    else:
        if (offset + iterator) <= business_num:
            offset += iterator
        else:
            offset = offset + (business_num % iterator)
        
            
    return offset
```
